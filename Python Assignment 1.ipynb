
# **String** **Concatenation**



```python
string1="Hello"
name=input("enter your name:")
Result=string1+" "+name
print(Result)
```

```python
string3=", welcome to Python programming"
Result=Result+string3
print(Result)
```

# **String Slicing and** **Indexing**

```python
print('first character:',Result[0])
```

```python
print('last character:',Result[-1])
```

```python
print('first five character:',Result[:5])
```

```python
print('Last 11 Character:',Result[-11:])
```

```python
print('Reverse character:',Result[::-1])
```

```python
print("Word 'Python':", Result[Result.index("Python"): Result.index("Python") + 6])
```

# **String Methods**




```python
strM = "Python beginner tutorial"
print('uppercase:',strM.upper())
```

```python
print('Lowercase:',strM.lower())
```

```python
print('Capitalize:',strM.capitalize())
```

```python
print('Count of t:',strM.count('t'))
```

```python
print('Replaced strM:', strM.replace('Python', 'Machine Learning'))
```

# **Tuples**


```python
Tuple1=(10,20,30)
Tuple2=(40,50,60)
t_combine=Tuple1+Tuple2
print('Concatenate Tuple:',t_combine)
```

```python
print('Repeat Tuple:',t_combine*3)
```

```python
print('3rd element:',t_combine[3])
```

```python
print('First three element:',t_combine[:3])
```

```python
print('Last three element:',t_combine[-3:])
```
